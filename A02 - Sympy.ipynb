$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np
sp.var("x")

def f(x):
    return 1260 + 117*x - 9*x**2
df = sp.diff(f(x),x)
pc = sp.solve(df , x)
ddf = sp.diff(df, x)
crit_2da_derivada = ddf.subs(x, pc[0])

if crit_2da_derivada < 0 :
    print("El punto crítico se encuentra en " + str(pc[0]) + " y es un máximo." )
elif crit_2da_derivada > 0 :
    print("El punto crítico se encuentra en " + str(pc[0]) + " y es un mínimo." )
else:
    print("El punto crítico se encuentra en " + str(pc[0]) + " y no se puede definir por criterio de segunda derivada")


import matplotlib.pyplot as plt
%matplotlib inline

xn = np.linspace(-30,43,1000)

plt.figure(figsize=(4,3))
plt.plot(xn, f(xn))
for c in pc:
    plt.plot(c, f(c), "*r", ms=10)
```

$f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
sp.var("x")
sp.var("y")

def f(x):
    return 5 + x + 4/x
df = sp.diff(f(x),x)
pc = sp.solve(df , x) 
ddf = sp.diff(df, x)

for p in pc:
    if p > 0:
        crit_2da_derivada = ddf.subs(x, p)
        if crit_2da_derivada < 0 :
            print("El punto crítico se encuentra en " + str(p) + " y es un máximo." )
        elif crit_2da_derivada > 0 :
            print("El punto crítico se encuentra en " + str(p) + " y es un mínimo." )
        else:
            print("El punto crítico se encuentra en " + str(p) + " y no se puede definir por criterio de segunda derivada")
            
            
xn = np.linspace(0.1,10,100)
plt.figure(figsize=(6,4))
plt.plot(xn, f(xn))
plt.plot(2, f(2), "*r", ms=10)            
```

$f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
sp.var("x")
sp.var("y")

def f(x):
    return 1/4*x**4 + 1/3*x**3 - x**2
df = sp.diff(f(x),x)
pc = sp.solve(df , x) 
ddf = sp.diff(df, x)
for p in pc:
    if p == 0:
            print("El punto crítico se encuentra en " + str(p) + " y es un máximo global." )
    elif p == -2:
            print("El punto crítico se encuentra en " + str(p) + " y es un mínimo global." )
    elif p == 1:
            print("El punto crítico se encuentra en " + str(p) + " y es un mínimo." )
    else:
        print("El punto crítico se encuentra en " + str(p) + " y no se puede definir por criterio de segunda derivada")     
            
xn = np.linspace(-3,1.7,100)
plt.figure(figsize=(6,4))
plt.plot(xn, f(xn))
plt.plot(-2, f(-2), "*y", ms=10)
plt.plot(0, f(0), "*b", ms=10)
plt.plot(1, f(1), "*g", ms=10)
```
